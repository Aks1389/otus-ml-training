# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
array_1d = np.zeros(10)
array_1d[4] = 1

array_2d = array_1d.reshape(2, 5)

print("1D Array:")
print(array_1d)

print("\n2D Array:")
print(array_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
array = np.arange(10, 50)
mirrored_array = array[::-1]
odd_elements = np.where(mirrored_array % 2 == 0)

print(array)
print(mirrored_array)
print(mirrored_array[odd_elements])
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
print(np.arange(9).reshape(3,3))
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
random_array = np.random.randint(1, 101, size=(4, 3, 2))

min_value = np.min(random_array)
max_value = np.max(random_array)

print("Random 4x3x2 Array:")
print(random_array)

print("\nMinimum Value:")
print(min_value)

print("\nMaximum Value:")
print(max_value)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
array1 = np.random.randint(0, 50, size=(6, 4))
array2 = np.random.randint(0, 50, size=(4, 3))

multiplication = np.dot(array1, array2)

print("array1")
print(array1)
print("\narray2")
print(array2)
print("\nmultiplication")
print(multiplication)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
random_array = np.random.randint(100, size=(7,7))

mean_value = np.mean(random_array)
std_value = np.std(random_array)
normalization = (random_array - mean_value) / std_value


print("random_array")
print(random_array)
print("\nmean_value")
print(mean_value)
print("\nstd_value")
print(std_value)
print("\nnormalization")
print(normalization)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pds
import matplotlib as mpl
import seaborn as sb
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = pds.read_csv("tips.csv")
```

### Посмотрите на первые 5 строчек

```python
tips[:5]
# или tips.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(tips.shape)
print(len(tips))
```

### Проверьте есть ли пропуски в данных

```python
missing_values = tips.isnull().sum()
print(missing_values)
```

### Посмотрите на распределение числовых признаков

```python
tips.describe()
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips["total_bill"].max()

print("Max in 'total_bill' column:")
print(max_total_bill)
```

### Найдите количество курящих людей

```python
tips[tips['smoker']=='Yes'].shape[0]
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
tips.groupby("day")["total_bill"].mean()
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
m = tips['total_bill'].median()
print(f"Median: {m}")
print("\nTotal_bill > Median")
print(tips[tips['total_bill'] > m])

print("\nСредний 'tip' в зависимости от 'sex'")
tips.groupby("sex")["tip"].mean()
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips["smoker"] = tips["smoker"].map({"Yes": 1, "No": 0})

print(tips.head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
tips["total_bill"].plot(kind="hist", bins=20, edgecolor="grey", alpha=0.7)

mpl.pyplot.xlabel("Total Bill")
mpl.pyplot.ylabel("Частота")
mpl.pyplot.title("Распределение значений признака total_bill")

mpl.pyplot.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
tips.plot.scatter(x='total_bill', y='tip')
mpl.pyplot.xlabel("Total Bill")
mpl.pyplot.ylabel("Tip")
mpl.pyplot.title("Scatter Plot: Total Bill vs Tip")
mpl.pyplot.show()
```

### Постройте pairplot

```python
sb.pairplot(tips, hue="tip")
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
mean_total_bill_by_day = tips.groupby("day")["total_bill"].mean()

mean_total_bill_by_day.plot(kind="line", marker="o", linestyle="-", color="blue", figsize=(8, 5))

mpl.pyplot.xlabel("Day")
mpl.pyplot.ylabel("Mean Total Bill")
mpl.pyplot.title("Line Plot: Mean Total Bill by Day")

mpl.pyplot.grid(True, linestyle="--", alpha=0.5)
mpl.pyplot.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
lunch_tips = tips[tips["time"] == "Lunch"]["tip"]
dinner_tips = tips[tips["time"] == "Dinner"]["tip"]

fig, axes = mpl.pyplot.subplots(1, 2, figsize=(12, 5))
lunch_tips.plot(kind="hist", bins=25, color="skyblue", edgecolor="black", ax=axes[0], alpha=0.7)
axes[0].set_title("Распределение Tips (Lunch)")
axes[0].set_xlabel("Tip")
axes[0].set_ylabel("Частота")

dinner_tips.plot(kind="hist", bins=25, color="salmon", edgecolor="black", ax=axes[1], alpha=0.7)
axes[1].set_title("Распределение Tips (Dinner)")
axes[1].set_xlabel("Tip")
axes[1].set_ylabel("Частота")

mpl.pyplot.tight_layout()
mpl.pyplot.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
df = pds.read_csv("tips.csv")

male_data = df[df["sex"] == "Male"]
female_data = df[df["sex"] == "Female"]

fig, axes = mpl.pyplot.subplots(1, 2, figsize=(14, 6))

scatter_male = axes[0].scatter(
    male_data["total_bill"], 
    male_data["tip"], 
    c=male_data["smoker"].map({"Yes": "red", "No": "blue"}),
    alpha=0.7, 
    edgecolor="black"
)
axes[0].set_title("Total Bill vs Tip (Male)")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Tip")
axes[0].grid(True, linestyle="--", alpha=0.5)

legend_labels_male = {"Yes": "Smoker", "No": "Non-Smoker"}
handles_male = [
    mpl.pyplot.Line2D([0], [0], marker="o", color="w", markerfacecolor="red", markersize=10, label="Smoker"),
    mpl.pyplot.Line2D([0], [0], marker="o", color="w", markerfacecolor="blue", markersize=10, label="Non-Smoker")
]
axes[0].legend(handles=handles_male, title="Smoker")

scatter_female = axes[1].scatter(
    female_data["total_bill"], 
    female_data["tip"], 
    c=female_data["smoker"].map({"Yes": "red", "No": "blue"}),
    alpha=0.7, 
    edgecolor="black"
)
axes[1].set_title("Total Bill vs Tip (Female)")
axes[1].set_xlabel("Total Bill")
axes[1].set_ylabel("Tip")
axes[1].grid(True, linestyle="--", alpha=0.5)

legend_labels_female = {"Yes": "Smoker", "No": "Non-Smoker"}
handles_female = [
    mpl.pyplot.Line2D([0], [0], marker="o", color="w", markerfacecolor="red", markersize=10, label="Smoker"),
    mpl.pyplot.Line2D([0], [0], marker="o", color="w", markerfacecolor="blue", markersize=10, label="Non-Smoker")
]
axes[1].legend(handles=handles_female, title="Smoker")

mpl.pyplot.show()
```
